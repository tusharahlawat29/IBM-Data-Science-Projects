## Step 1: Import Required Libraries

We begin by importing essential Python libraries for data analysis and visualization:
- `pandas` for data handling
- `seaborn` and `matplotlib.pyplot` for visualizations



```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## Step 2: Load the Dataset

We load the famous Iris dataset using Seaborn's built-in `load_dataset` method.


```python
iris_df = sns.load_dataset('iris')
```

## Step 3: Preview the Dataset

Let's view the first few rows of the dataset using `.head()`.


```python
iris_df.head()
```

## Step 4: Basic Information about the Dataset

We check for:
- Data types of each column  
- Null values  
- Number of entries  
using `.info()`


```python
iris_df.info()
```

## Step 5: Statistical Summary

We use `.describe()` to understand the central tendency, dispersion, and distribution of numerical features.


```python
iris_df.describe()
```

## Step 6: Univariate Analysis - Histograms

We plot histograms to visualize the distribution of each numerical feature in the dataset.


```python
iris_df.hist(figsize=(10, 8), bins=20)
plt.suptitle('Histograms of Iris Features')
plt.show()
```

## Step 7: Count of Each Species

We use a countplot to check the distribution of the target variable `species`.


```python
sns.countplot(x='species',hue='species', data=iris_df, palette='Set2', legend=False)
plt.title('Count of Each Iris Species')
plt.show()
```

## Step 8: Bivariate Analysis - Pairplot

We use a pairplot to visualize relationships between each pair of numerical features, colored by species.


```python
sns.pairplot(iris_df, hue='species', palette='Set1')
plt.suptitle('Pairwise Scatterplots Colored by Species', y=1.02)
plt.show()
```

## Step 9: Boxplots of Each Feature by Species

We plot boxplots for each feature to compare distributions between the three Iris species.


```python
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 10))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(x='species',hue='species', y=feature, data=iris_df, palette='Set3')
    plt.title(f'Boxplot of {feature} by Species')
plt.tight_layout()
plt.show()
```

## Step 10: Correlation Heatmap

We create a heatmap to examine the correlations between numerical features.


```python
corr = iris_df.select_dtypes(include='number').corr()

plt.figure(figsize=(8, 6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap of Iris Features')
plt.show()
```

## Step 11: Summary and Insights

- **Petal length** and **petal width** have a strong positive correlation (~0.96).
- **Sepal width** is weakly or negatively correlated with other features.
- **Setosa** species is clearly separable based on petal features.
- **Versicolor** and **Virginica** have overlapping feature distributions.
- These visualizations help us understand which features might be important for building predictive models.

This concludes our exploratory data analysis of the Iris dataset.
