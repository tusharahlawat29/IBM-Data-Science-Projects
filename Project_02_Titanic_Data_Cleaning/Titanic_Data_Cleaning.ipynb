```python
import pandas as pd

url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

titanic_df = pd.read_csv(url)

titanic_df.head()
```

### Loading Titanic Dataset

- We load the dataset into a pandas DataFrame.
- Display the first few rows using `.head()` to understand columns and sample data.


```python
missing_counts = titanic_df.isnull().sum()
missing_counts
```

### Checking Missing Data

- Using `.isnull().sum()`, we count missing values (NaNs) per column.
- This helps us decide which columns need cleaning.


```python

titanic_df['Age']= titanic_df['Age'].fillna(titanic_df['Age'].median())


titanic_df['Embarked']= titanic_df['Embarked'].fillna(titanic_df['Embarked'].mode()[0])


titanic_df.drop('Cabin', axis=1, inplace=True)

titanic_df.isnull().sum()

```

### Handling Missing Data

- Filled missing 'Age' values with the median age.
- Filled missing 'Embarked' values with the most frequent category.
- Dropped the 'Cabin' column due to excessive missing data.
- Verified that the dataset no longer contains missing values in important columns.


```python
survivors = titanic_df[titanic_df['Survived'] == 1]

survivors.head()
```
